### Chapter 5: Test the following hypotheses <a class="anchor" id="chapter5"></a>

#### Section 5.1: Average user ratings of the Xbox One and PC platforms are the same <a class="anchor" id="section_5_1"></a>


The dataset has filetered for the XOne anf PC paltforms and convereted the user_score to the float dtype.After, the missing values=0 rows has dropped.

```python
XBOX_platform=relevant_years.query('platform=="XOne"')

PC_platform=relevant_years.query('platform=="PC"')
XBOX_platform['user_score']=XBOX_platform['user_score'].astype('float')
XBOX_platform=XBOX_platform.query('user_score !=0')
PC_platform['user_score']=PC_platform['user_score'].astype('float')
PC_platform=PC_platform.query('user_score !=0')
```

```python
display(XBOX_platform.head(),PC_platform.head())
```

__Formulation of the null and alternative hypotheses__

__The Null Hypothsis Formulation__


The two-sided test is used which mean we don not care the value we will get is lower or higher. The sample data of total population is selected and to accpet the null hypothesis (H0), we stated that average user ratings of the Xbox One and PC platforms are  the same. 

__The alternative Hypothis__

If the H0 is rejected and can conclude that the average user rating is  not same. 

```python
alpha=0.05

results=st.ttest_ind(XBOX_platform['user_score'],PC_platform['user_score'])

print('p-value:', results.pvalue)

if(results.pvalue<alpha):
    print('We reject null hypothesis')
else:
    print("We can't reject the null hypothesis")
```

The test of our selected sample of total dataset describe that H0 is rejected so that means the average user ratings of the Xbox One and PC platforms are not the same.

The significance level of 0.05 is chosed that's indicates a 5% risk of concluding that a difference exists when there is no actual difference.

#### Section 5.2: Average user ratings for the Action and Sports genres are different. <a class="anchor" id="section_5_2"></a>

Firstly, the genre data of action and Sports will be filtered and 

```python
action_genre=relevant_years.query('genre=="Action"')
sports_genre=relevant_years.query('genre=="Sports"')

action_genre['user_score']=action_genre['user_score'].astype('float')
sports_genre=sports_genre.query('user_score !=0')

sports_genre['user_score']=sports_genre['user_score'].astype('float')
sports_genre=sports_genre.query('user_score !=0')
```

```python
alpha=0.05

results=st.ttest_ind(action_genre['user_score'],sports_genre['user_score'])

print('p-value:', results.pvalue)

if(results.pvalue<alpha):
    print('We reject null hypothesis')
else:
    print("We can't reject the null hypothesis")
```

The test of our selected sample of total dataset describe that H0 is rejected so that means the average user ratings of the action and sports user rating are not the same.
